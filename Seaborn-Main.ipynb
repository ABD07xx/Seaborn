# Seaborn 

Seaborn is a popular Python data visualization library built on top of the Matplotlib library. It provides a high-level interface for creating beautiful and informative statistical graphics.

Seaborn is designed to work well with pandas data structures and supports a wide range of data visualization techniques, including univariate and bivariate plotting, categorical plotting, matrix plotting, and regression analysis. It also includes built-in color palettes and themes to enhance the visual aesthetics of the plots.

Seaborn makes it easy to create complex and polished visualizations with minimal code. Seaborn also provides an API for customizing the visual properties of plots and supports various output formats such as PDF, PNG, and SVG.

# Univariate data <a id='univariate'></a>

Univariate data refers to a type of data that consists of only one variable or feature. In other words, univariate data is data that is described by a single characteristic or attribute. For example, the height of students in a classroom or the weight of apples in a basket are both examples of univariate data.

```python
df = sns.load_dataset("iris")
df.head()
```

```python
df['sepal_length'].dtype
```

# Distplot <a id='dplot'></a>
Distplot pr Distribution Plot allows us to visualize the distribution of a set of univariate data. It can create a <b>`histogram`</b> or <b>`kernel density estimate/KDE`</b> plot, or both, in a single plot. A histogram is a graph of the frequency distribution of data, while a kernel density estimate plot is a non-parametric estimate of the probability density function of a random variable. 

```python
sns.distplot(df['sepal_length'])
plt.show
```

```python
#Hist = Hist bars
sns.distplot(df['sepal_length'],hist=False)
plt.show
```

```python
#Kde - Trend
sns.distplot(df['sepal_length'],kde=False)
plt.show
```

## Changing various attributes of Distplot <a id='change'></a>

```python
sns.set_context('poster')
sns.distplot(df['petal_length'],hist=False)
plt.show
```

```python
sns.set_style('whitegrid')
sns.set_context('notebook')
sns.distplot(df['petal_length'],hist=False)
plt.show
```

```python
sns.set_style('darkgrid')
sns.set_context('notebook')
sns.distplot(df['petal_length'],hist=False)
plt.show
```

```python
sns.set_theme(context='poster',
    style='darkgrid')
sns.distplot(df['petal_length'],hist=False)
plt.show()
```

# Bivariate data <a id='bdata'></a>

Bivariate data refers to a type of data that involves two variables or features, and their relationship or association with each other. For example, the height and weight of a group of individuals or the temperature and humidity of a location are both examples of bivariate data. 

We can broadly define our data in two main categories <b>`Numerical`</b> and <b>`Categorical`</b>. Using Seaborn We can plot a graph when: 
1. Both variables are <b>`Numerical`</b> or we can say a <b>`Numerical to Numerical`</b> relationship
2. <b>`Numerical to Categorical`</b> relationship
3. <b>`Categorical to Categorical`</b> relationship

# Relplot  <a id='relplot'></a>

Relplot allows us to create a figure with one or more subplots to visualize the relationship between two numerical variables. t provides a flexible interface to create a wide range of visualizations, including scatter plots, line plots, bar plots etc. This can be useful for identifying patterns and trends in the data that may be obscured in an overall view.

```python
df = sns.load_dataset("tips")
df.head(5)
```

```python
sns.set_context("notebook")
sns.relplot(data=df, x="total_bill", y="tip" ,kind='scatter')
```

```python
sns.set_context("notebook")
sns.relplot(data=df, x="total_bill", y="tip" ,kind='line')
```

```python
sns.set_context("notebook")
sns.relplot(data=df, x="total_bill", y="tip" ,hue='day')
```

```python
df['day'].describe()
```

<a id='pm'></a>
<b>`hue`</b> parameter allows us to map a third categorical variable to the color of the plot markers or the color of the plot lines in a line plot. This can be useful for visualizing the relationship between two variables while also displaying the distribution of a third variable. 

```python
sns.relplot(data=df, x="total_bill", y="tip",col='day',hue='sex')
```

<b>`col_wrap`</b> parameter allows us to control the number of columns in the facet grid. By default, relplot will try to create a plot with a single row and as many columns as necessary to fit all the subplots as seen above. However, if you set col_wrap to a number greater than 1, the plot will be wrapped into multiple rows, each with the specified number of columns. This can be useful for creating more compact plots when you have a large number of subplots.

```python
sns.relplot(data=df, x="total_bill", y="tip",col='day',hue='sex',col_wrap=2)
```

```python
sns.relplot(data=df, x="total_bill", y="tip",col='day',hue='sex',kind='line',col_wrap=2)
```

<b>`kind`</b> is a parameter that allows you to specify the type of plot to create.


```python
sns.relplot(data=df, x="size", y="tip",kind="line")
```

```python
sns.relplot(data=df, x="total_bill", y="tip",col='day',hue='smoker',col_wrap=2)
```

# Barplot <a id='barplot'></a>
Barplot is a relation between a categorical variable and a continuous variable. A barplot is a type of plot that shows an estimate of a central tendency (usually the mean) for a numeric variable with the height of each rectangle and the error bars indicating the confidence interval around the estimate. It is useful for comparing the values of a numeric variable for different categories

```python
df_4 = sns.load_dataset("titanic")
df_4.head()
```

```python
sns.set_theme(
    context="notebook",
    style="whitegrid"
)
sns.barplot(data=df_4,y="survived",x="sex",saturation=1,palette="Set2")
```

```python
sns.set_theme(
    context="notebook",
    style="whitegrid"
)
sns.barplot(data=df_4,y="survived",x="sex",hue='class',saturation=1,palette="Set2")
```

```python
sns.barplot(x='age',y='embark_town',data=df_4,
            palette="PuRd")
```

# Countplot <a id='cplot'></a>

Countplot is a type of bar plot that displays the count of observations in each categorical bin using bars. It is useful for visualizing the distribution of a categorical variable. 

```python
sns.countplot(data=df_4,x = 'class')
```

```python
# Adding count number on top
ax = sns.countplot(x = "class", data = df_4, palette = "Blues");
ax.bar_label(container = ax.containers[0])
```

```python
cnt = sns.countplot(data=df_4,x="class",hue="who",palette='magma')
#cnt.bar_label(cnt.containers[0])
#cnt.bar_label(cnt.containers[1])
#cnt.bar_label(cnt.containers[2])
for i in cnt.containers:
    cnt.bar_label(i)

plt.title("Surviors")
```

# Violinplot <a id='Violinplot'></a>

A violin plot is a type of plot that shows the distribution of a numeric variable for each category using a kernel density estimate. It is useful for comparing the shape of the distribution of a numeric variable for different categories

```python
sns.set_context("poster")
sns.set_style("whitegrid")
df_3 = sns.load_dataset('tips')
sns.violinplot(data=df_3,x = "day", y = "total_bill",saturation=1)
```

```python
sns.set_context("notebook")
sns.set_style("white")
plt1 = sns.violinplot(data=df_3,x = "day", y = "total_bill",hue='sex',saturation=1,palette="Set1")
```

```python
#Splitting the data in Violin Plot
sns.violinplot(data=df,x="day",y="total_bill",hue='sex',split=True,saturation=1,palette="Set1")
```

# Regplot <a id='regplot'></a>

Regplot is a function in Seaborn that creates a scatter plot with a linear regression line overlaid on it. This plot is useful for visualizing the relationship between two numerical variables and to determine if a linear regression model is a good fit for the data. Regplot accepts the x and y variables in a variety of formats including simple numpy arrays, pandas Series objects, or as references to variables in a pandas DataFrame

```python
ax = sns.regplot(x="total_bill",y="tip",data=df)
```

```python
## We can also add some jitters around 
sns.regplot(x="total_bill", y="tip", data=df)
```

```python
sns.regplot(x="total_bill", y="tip", data=df, x_jitter=1, ci = None)
```

# lmplot <a id='lmplot'></a>

<b>`lmplot`</b> is a function that creates a scatter plot with a linear regression line overlaid on it, and allows us to plot multiple subsets of the data. This plot is useful for visualizing the relationship between two numerical variables, and for exploring the effects of additional categorical variables on that relationship.

```python
sns.lmplot(x="total_bill", y="tip", col="day", hue="day",
               data=df, col_wrap=2, height=5)
```

# HeatMap <a id='heatmap'></a>
A heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colours.

```python
flights=sns.load_dataset("flights")
flights = flights.pivot( "month","year", "passengers")
flights
```

```python
sns.heatmap(flights,cmap="RdBu")
```

```python
sns.heatmap(flights,cmap="Greens",linewidths=.5) # more friendly
```

# Pairplot <a id='pairplot'></a>

It creates a grid of Axes such that each numeric variable in data will be shared across the y-axes across a single row and the x-axes across a single column.
The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.
Pair Plots are a really simple way to visualize relationships between each variable

```python
df = sns.load_dataset('iris')
sns.set_style("ticks")
sns.pairplot(df,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```
